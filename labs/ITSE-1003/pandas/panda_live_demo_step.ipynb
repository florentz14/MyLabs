# Panda Live Demo Step

Notebook version of `panda_live_demo_step.py` with one step per cell.

```python
from pathlib import Path

import pandas as pd
from IPython.display import display

# Notebook lives in labs/ITSE-1003/pandas/, so data/ is one level up.
data_dir = Path.cwd().parent / "data"

# step 1: load student rows from CSV
df = pd.read_csv(data_dir / "students.csv")
df.columns = df.columns.str.strip()
```

```python
# step 2: print the full DataFrame
display(df)
```

```python
# step 3: show first 5 rows
display(df.head())
```

```python
# step 4: show last 5 rows
display(df.tail())
```

```python
# step 5: show DataFrame info
print("\nInfo:")
df.info()
```

```python
# step 6: show numeric summary statistics
display(df.describe())
```

```python
# step 7: show column names
display(df.columns)
```

```python
# step 8: select one column
display(df["Name"])
```

```python
# step 9: select multiple columns
display(df[["Name", "Age", "Major"]])
```

```python
# step 10: select rows using loc
display(df.loc[df["Age"] > 30])
```

```python
# step 11: select rows using slice
display(df[0:3])
```

```python
# step 12: select one row using iloc
display(df.iloc[0])
```

```python
# step 13: filter Age > 30
display(df[df["Age"] > 30])
```

```python
# step 14: filter Age > 20 using loc
display(df.loc[df["Age"] > 20])
```

```python
# step 15: same filter using iloc
display(df.iloc[(df["Age"] > 20).to_numpy().nonzero()[0]])
```

```python
# step 16: first three rows again
display(df[0:3])
```

```python
# step 17: grade letter A
display(df.loc[df["grade_letter"].str.strip() == "A"])
```

```python
# step 18: grade > 70
display(df.loc[df["grade"] > 70])
```

```python
# step 19: add Passed column
df["Passed"] = df["grade"] >= 70
display(df)
```

```python
# step 20: add Status column
df["Status"] = df["grade"].apply(lambda x: "Passed" if x >= 70 else "Failed")
display(df)
```

```python
# step 21: sort by Age descending
display(df.sort_values(by="Age", ascending=False))
```

```python
# step 22: sort by Age ascending
display(df.sort_values(by="Age", ascending=True))
```

```python
# step 23: group by Major and count rows
display(df.groupby("Major").size())
```

```python
# step 24: average grade per major
display(df.groupby("Major")["grade"].mean())
```

```python
# step 25: highest grade
display(df[df["grade"] == df["grade"].max()])
```

```python
# step 26: lowest grade
display(df[df["grade"] == df["grade"].min()])
```

```python
# step 27: highest GPA
display(df[df["GPA"] == df["GPA"].max()])
```

```python
# step 28: combined conditions
display(df[(df["Age"] > 30) & (df["Major"] == "Computer Science")])
```

```python
# step 29: save updated CSV
print("\nSave the updated DataFrame to a new CSV file:")
df.to_csv(data_dir / "students_updated.csv", index=False)
```
